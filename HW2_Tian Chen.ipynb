```python
#set selenium driver
driver_path = Service('/Users/tinchen/Desktop/WashU/Python/pythoncamp2023/HW/chromedriver')
```

```python
url = 'https://www.presidency.ucsb.edu/documents/app-categories/presidential/spoken-addresses-and-remarks?page=' 
n = 0
data_list = []
for n in range(0,109):
    print("working on page {} of 108 pages".format(n+1))
    driver = webdriver.Chrome(service = driver_path)
    page = n
    html = url + str(page) #go to next page
    #use selenium driver 
    driver.get(html)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") # scroll to bottom
    time.sleep(5)
    link = driver.page_source
    driver.close()
    soup = BeautifulSoup(link,"html.parser")
    results = soup.find_all('div', class_='col-sm-8')

    for res in results:
        #polite sleep
        time.sleep(random.randint(1, 5))
        data = {}
        data["Date"] = res.find('span', class_='date-display-single').text.strip() #Date
        data["Title"] = res.find('p').text.strip() #Title
        # handle interior page
        interior = 'https://www.presidency.ucsb.edu/' + res.find('a')['href']
        req = urllib.request.urlopen(interior)
        interior_page = req.read()

        # Parse interior page
        interior_soup = BeautifulSoup(interior_page, 'html.parser')
        data["President"] = interior_soup.find('h3', class_="diet-title").a.text.strip() #President's name
        paragraphs = interior_soup.find_all('div', class_="field-docs-content")
        address = ''
        for paragraph in paragraphs:
            address += paragraph.text.strip() + '\n'
        data["AddressRemark"] = address.strip() #address content 
        data["Link"] = interior #interior link
        citation = interior_soup.find('p', class_="ucsbapp_citation")#citation page
        if citation:
            data["Citation"] = citation.text.strip()
        else:
            data["Citation"] = "No citation available"
        data_list.append(data)
        
        #polite sleep
        time.sleep(random.randint(3, 5))
        n += 1
```

```python
# Print the extracted data
for num in range(0, 5):
    print(data_list[num])
print
```
